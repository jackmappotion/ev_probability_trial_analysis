```python
# kc = KellyCriterion
# gkc = general_kc
# rkc = rough_kc


def calc_gkc(a, b, p):
    f = (p / a) - ((1 - p) / b)
    return f


def calc_rkc(b, p):
    """
    calc_gkc with a = 1
    """
    f = (p) - ((1 - p) / b)
    return f
```

```python
def get_calc_b(ev):
    def calc_b(p):
        v = ev / p
        b = v - 1
        return b

    return calc_b


calc_b = get_calc_b(ev=1.2)
```

```python
p = 0.1
f = calc_rkc(calc_b(p), p)
round(1 / f)
```

```python
p = 0.3
f = calc_gkc(0.1, 0.3, p)
round(1 / f)
```
